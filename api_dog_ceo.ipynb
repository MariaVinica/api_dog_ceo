<h1> Dog CEO API connection </h1>

```python
# Imports necesarios:
# Json, para trabajar datos que se encuentren en este formato
# Request, para realizar solicitudes HTTP
# Matplotlib y Skimage para trabajar con imágenes y poder mostrarlas

import json
import requests
import matplotlib.pyplot as plt
from skimage import io
```

```python
print("Boosting the mood with random puppies. For that, you need to know about dog breeds🤗")

while True:
    breed = input("Breed: ")
    url = f"https://dog.ceo/api/breed/{breed}/images/random"

    answer = requests.get(url)
    dog = json.loads(answer.text)

    if answer.status_code == 200 and dog["status"] == "success":
        image_url = dog["message"]
        image = io.imread(image_url)
        plt.axis('off')
        mappable = plt.imshow(image)
        plt.show()
        break
    else:
        print("Ups😞! Try it again!")
```
